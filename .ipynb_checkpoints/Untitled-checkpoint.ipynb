```python
import requests
from collections import Counter

API_URL = "https://data.ademe.fr/data-fair/api/v1/datasets/dpe02neuf/lines"

# On teste plusieurs offsets pour voir la distribution réelle
offsets_a_tester = [ 5000000]

for offset in offsets_a_tester:
    response = requests.get(API_URL, params={"size": 1000, "from": offset})
    data     = response.json()
    lignes   = data.get("results", [])
    
    if not lignes:
        print(f"Offset {offset:>8,} : aucune donnee")
        continue
    
    classes = Counter(l.get("etiquette_dpe", "null") for l in lignes)
    classes_str = " | ".join(f"{k}:{v}" for k, v in sorted(classes.items()))
    print(f"Offset {offset:>8,} : {classes_str}")
```
